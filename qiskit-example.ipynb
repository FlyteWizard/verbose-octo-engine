## Qiskit Example - Apply Hadamard Gates and Draw Circuit

A Qiskit program that applies a Hadamard gate to `n` qubits initialized at $|0\rangle$, measured, and drawn as a quantum circuit.

### Create a Qiskit Program

Import the Qiskit Python modules. Create a Quantum Circuit with `n` qubits apply a Hadamard gate to each qubit, and measure the qubits. 

```python
from qiskit import Aer, assemble, transpile
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.visualization import plot_histogram
```

```python
quantum_register = QuantumRegister(3, 'q')
quantum_circuit = QuantumCircuit(quantum_register)

quantum_circuit.h(quantum_register[0])
quantum_circuit.h(quantum_register[1])
quantum_circuit.h(quantum_register[2])

quantum_circuit.measure_all()
```

### Draw a Quantum Circuit

Draw the quantum circuit. Use the `mpl` output format to draw the circuit using Matplotlib. Use the `style` parameter to set the style to `bw` for black and white.

```python
quantum_circuit.draw(output='mpl', style='bw', scale=0.75)
```

### Plot a Quantum Circuit

Use the `plot_histogram` function to plot the quantum circuit.

```python
aer_sim = Aer.get_backend('aer_simulator')
transpiled_quantum_circuit = transpile(quantum_circuit, aer_sim)

quantum_object = assemble(transpiled_quantum_circuit)
results = aer_sim.run(quantum_object).result()
counts = results.get_counts()

plot_histogram(
    counts, 
    title='Hadamard Gate on 3 Qubits', 
    color='midnightblue', 
    figsize=(9, 6.5)
)
```
